```python
data_triathlon = pd.read_csv('data/medal_country_year_triathlon.csv')
data_triathlon
```

```python
def calculer_distance(longitudeA,latitudeA,yearA,longitudeB,latitudeB,yearB):
    K = 1
    result = sqrt((longitudeA-longitudeB)^2 + (latitudeA-latitudeB)^2) + K * abs(yearA-yearB)
    return result
```

```python
cls_triathlon = KNeighborsClassifier(n_neighbors=5)
X_triathlon = pd.DataFrame()
X_triathlon['longitude'] = data_triathlon['longitude']
X_triathlon['latitude'] = data_triathlon['latitude']
X_triathlon['year'] = data_triathlon['year']
y_triathlon = data_triathlon['value']
```

```python
X_triathlon_train, X_triathlon_val, y_triathlon_train, y_triathlon_val = train_test_split(X_triathlon, y_triathlon, train_size=0.66)
cls_triathlon.fit(X_triathlon_train, y_triathlon_train) 
labels_triathlon = cls_triathlon.predict(X_triathlon_val)
right = 0
false = 0
a = y_triathlon_val.tolist()
y_triathlon_val_cpy = []
for i in range(len(labels_triathlon)):
    if labels_triathlon[i] ==  a[i]:
        right += 1
    else:
        false += 1
print('eff = ', right , false)
```

```python
n_neighbors_list = np.unique(np.round(np.geomspace(1, 500, 100)).astype(int))
gen = utils_knn_2.knn_simple_validation(X_triathlon_train, y_triathlon_train, X_triathlon_val, y_triathlon_val, n_neighbors_list)
df = pd.DataFrame(gen, columns=["# neighbors", "accuracy", "degrés de liberté"])
```

```python
sp1 = sns.lineplot(x="degrés de liberté", y="accuracy", data=df)
sp1.set(xscale="log")
plt.savefig("new images/triathlon_knn_simple_validation_medal_by_longlat_accuracy_1")
```

```python
sp2 = sns.lineplot(x="# neighbors", y="accuracy", data=df)
sp2.set(xscale="log")
plt.savefig("new images/triathlon_knn_simple_validation_medal_by_longlat_accuracy_2")
```

```python
Kopt = df.groupby("# neighbors").mean().accuracy.idxmax()
Kopt
```

```python
cls = KNeighborsClassifier(n_neighbors=Kopt)
c=1
for year in data_triathlon["year"].unique():
    plt.figure(c)
    plt.subplot().set_title(year*4 + 1980)
    x_visual = pd.DataFrame()
    data_year = data_triathlon[data_triathlon["year"] == year]
    
    x_visual['longitude'] = data_year['longitude']
    x_visual['latitude'] = data_year['latitude']
    y_visual = data_year['value']
    cls.fit(x_visual, y_visual)
    utils_knn.plot_clustering(x_visual, y_visual)
    utils_ald.add_decision_boundary(cls)
    c += 1
```
